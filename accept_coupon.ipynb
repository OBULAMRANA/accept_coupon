### 3. Compare the acceptance rates between those drivers who
##### Generate a dataframe with Coupon type Bar and compare who went less than 3 times vs others
##### get the dateset with coupon = bar & get who visited less than 3 times( from data "1~3", "less1", "never")
##### get the dateset with coupon = bar & get who visited more  than 4 times( from data "4~8", "gt8")

```python

data_bar=data[data['coupon']=='Bar']

data_bar_less_3 = data_bar[ (data_bar['Bar'] == "1~3")  | (data_bar['Bar'] == "less1") | (data_bar['Bar'] == "never") ]
bar_went_less_3_counts = data_bar_less_3['Y'].value_counts()

data_bar_above_4 = data_bar[ (data_bar['Bar'] == "4~8")  | (data_bar['Bar'] == "gt8") ]
bar_went_above_4_counts = data_bar_above_4['Y'].value_counts()

bar_passenger_went_total_counts = data_bar['Y'].value_counts()

print("Bar went less than 3 times:")
print(len(data_bar_less_3))
print("Bar went more than 4 times:")
print(len(data_bar_above_4))
print("Total Bar Accepted Count:")
print(len(data_bar))

print("% People went less than 3 times")
print(len(data_bar_less_3)/len(data_bar) * 100)

print("% People went more than 4 times")
print(len(data_bar_above_4)/len(data_bar) * 100)
#data_bar_counts.plot(kind='pie')

print("Passengers went more than 3 times 89% vs 4 times 9%")

data_bar_less_3_acceptance_rate = data_bar_less_3['Y'].value_counts(normalize=True) * 100
print("Acceptance rate of bar cupons with less than 3 time went:")
print(data_bar_less_3_acceptance_rate)

data_bar_above_4_acceptance_rate = data_bar_above_4['Y'].value_counts(normalize=True) * 100
print("Acceptance rate of bar cupons with more than 4 time went:")
print(data_bar_coupon_counts_acceptance_rate)

print("Conclution/Findings: 37% went less than 3 times & 41% went more than 4 times")
```

### Compare the acceptance rate
#####between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?

```python

data_bar_once_month=data[(data['coupon']=='Bar') & (data['Y']==1) ]
data_bar_age_below_25 = data_bar_once_month[ (data_bar_once_month['age'].str.contains('21', regex=True))]
data_bar_age_above_25 = data_bar_once_month[ ~data_bar_once_month['age'].str.contains('21', regex=True)]
print("Total Bar Accepted Count:")
print(len(data_bar))
print("Drivers below 25:")
print(len(data_bar_age_below_25))
print("Drivers above 25:")
print(len(data_bar_age_above_25))
print("% of drivers below 25  go to bar more than once:")
print(len(data_bar_age_below_25)/len(data_bar_once_month) * 100)
print("% of drivers above 25  go to bar more than once:")
print(len(data_bar_age_above_25)/len(data_bar_once_month) * 100)
print("\n Conclusion/Findings: 40% of more drivers go to the bar above the age 25 compared to below 25 age. \n")

df_accepted = { 'drivers':['under-25', 'above-25'],
               'percentage': [29, 70]}
plt.title("Accepted Rate Comparision")
plt.xlabel("Drivers")
plt.ylabel("Percentage")
plt.bar(df_accepted['drivers'], df_accepted['percentage'], color=['orange', 'blue'])
plt.show()
```

###Compare the acceptance rates between those drivers who:

###### - go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*'
###### maritalStatus != Widowed

```python
data_bar=data[data['coupon']=='Bar']
# Go to bars more than once a month
data_bar_morethan1 = data_bar[(data_bar['Bar'] == "1~3")  | (data_bar['Bar'] == "4~8")  | (data_bar['Bar'] == "gt8") ]

# Go to bars more than once a month with no kinds and no widows
data_bar_morethan1_nkid_nwidow= data_bar_morethan1[ ~data_bar_morethan1['passanger'].str.contains("Kid") & ~data_bar_morethan1['maritalStatus'].str.contains("Widowed") ]
print("Go to bars more than once a month with no kinds and no widows:")
print(len(data_bar_morethan1_nkid_nwidow))


# go to bars more than once a month and are under the age of 30  # that means age 26,21, below21. Since 21 contains below21, so no need to add below21
#data_bar_morethan1_under30 = data_bar_morethan1[ (data_bar_morethan1['age'].str.contains('below21', regex=True)) | (data_bar_morethan1['age'].str.contains('21', regex=True)) | (data_bar_morethan1['age'].str.contains('26', regex=True))]
data_bar_morethan1_under30 = data_bar_morethan1[ (data_bar_morethan1['age'].str.contains('21', regex=True)) | (data_bar_morethan1['age'].str.contains('26', regex=True))]
print("Go to bars more than once a month and are under the age of 30:")
print(len(data_bar_morethan1_under30))

#go to cheap restaurants more than 4 times a month and income is less than 50K
#Preparing the data with income below50K and gt4 for "4`8" & "gt8"


#data_bar['income'] =data_bar.loc[index_of(data_bar['income'].str.contains('$37500 - $49999', regex=True)), 'income'] = "below50k"

data_bar.loc[data_bar['income'].str.contains('$37500 - $49999', regex=True), 'income'] = "below50k"
data_bar.loc[data_bar['income'].str.contains('$12500 - $24999', regex=True), 'income'] = "below50k"
data_bar.loc[data_bar['income'].str.contains('$25000 - $37499', regex=True), 'income'] = "below50k"

data_R4_I50 =data_bar[data_bar['income'].str.contains('below50k', regex=True) & (data_bar['RestaurantLessThan20'].str.contains('4~8', regex=True)) | (data_bar['RestaurantLessThan20'].str.contains('gt8', regex=True))]

print("Go to cheap restaurants more than 4 times a month and income is less than 50K:")
print(len(data_R4_I50 ))

print("% more than once vs other criterias \n")
print("  Go to bars more than once a month with no kinds and no widows:")
print(len(data_bar_morethan1_nkid_nwidow)/len(data_bar_morethan1) * 100)
print("\n  Go to bars more than once a month and are under the age of 30:")
print(len(data_bar_morethan1_under30)/len(data_bar_morethan1) * 100)
print("\n. Go to cheap restaurants more than 4 times a month and income is less than 50K:")
print(len(data_R4_I50)/len(data_bar_morethan1) * 100)


print("\n")
bar_morethan1_under30_acceptance_rate = data_bar_morethan1_under30['Y'].value_counts(normalize=True) * 100
print("Acceptance rate Go to bars more than once a month and are under the age of 30:")
print(bar_morethan1_under30_acceptance_rate)

print("\n")
bar_morethan1_nkid_nwidow_acceptance_rate = data_bar_morethan1_nkid_nwidow['Y'].value_counts(normalize=True) * 100
print("Acceptance rate Go to bars more than once a month with no kinds and no widows:")
print(bar_morethan1_nkid_nwidow_acceptance_rate)

R4_I50_acceptance_rate = data_R4_I50['Y'].value_counts(normalize=True) * 100
print("Acceptance rate restaurants more than 4 times a month and income is less than 50K::")
print(R4_I50_acceptance_rate )

print("\n")

print("Conclution/Findings: 72% accepted the cupons under age 30, 71% accepted  with no kids and windows & 45% accepted with more than 4 times with below 50K salary")

df_accepted = { 'drivers':['under-30', 'no-kids-widow', 'M4-Income50'],
               'percentage': [72,71,45]}
plt.title("Accepted Rate Comparision")
plt.xlabel("Drivers")
plt.ylabel("Percentage")
plt.bar(df_accepted['drivers'], df_accepted['percentage'], color=['red', 'green', 'blue'])
plt.show()

```
